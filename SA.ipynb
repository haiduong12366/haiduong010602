```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def Simulated_Annealing(current,temps):
    current=current
    x=[]
    y=[]
    x.append(current)
    y.append(value(current))
    for i in range (10000):
        nextnode = random.uniform(-10,10)
        deltaE=value(current)-value(nextnode)
        if(deltaE>0):
            current=nextnode
            x.append(current)
            y.append(value(current))
        else:
            acceptance_probability = math.exp(deltaE/temps)
            if(random.random()<acceptance_probability):
                current=nextnode
                x.append(current)
                y.append(value(current))
        temps = temps*0.99 #cooling_rate = 0.99
    return current,x,y

def value(x):
    return x**2

def y_coordinates(x):
    y=[]
    for i in x:
        y.append(value(i))
    return y

temps=100
current = random.uniform(-10,10)
current,x,y=Simulated_Annealing(current,temps)

X = np.linspace(-10,10,10000)
Y = y_coordinates(X)

print(current)
plt.plot(X,Y)
plt.plot(x,y)
```
